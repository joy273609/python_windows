```python
#requests為module
import requests
url="https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate%20desc&format=JSON"
#使用requests裡面的實體方法get()取得url
respone=requests.get(url)
type(respone)
```

```python
#使用實體方法method()
data=respone.json()
type(data)
```

```python
for items in data["records"]:
    print(items)
```

```python
#使用list串列，是有序且可能包含重複項的列表。
sitenames=[]
for items in data["records"]:
    sitenames.append(items["sitename"])
print(sitenames)
print("--------------------------------------")
print(len(sitenames))
```

```python
#使用set集合，集合內的資料不重複，且無法保證順序。
sitenames=set()
for items in data["records"]:
    sitenames.add(items["sitename"])
print(sitenames)
print("--------------------------------------")
print(len(sitenames))
print("--------------------------------------")
#將set轉乘list才能取得裡面的index及資料
sitenames=list(sitenames)
sitenames
```
